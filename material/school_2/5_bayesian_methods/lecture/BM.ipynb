## $$\textbf{What are Bayesian Methods?}$$
<hr>

* $\t{Bayesian Methods are a statistical approach that applies the Bayes' theorem}$

$\quad\quad\quad\t{to update the probability of a hypothesis as more evidence becomes available.}$

* $\t{It contrasts with classical statistics by incorporating prior knowledge alongside new data.}$

## $$\textbf{Probability rules}$$
<hr>



* $\textit{Product rule}: \text{conjunction of two events } A \text{ and } B$

$$P(A \cap B ) = P(A,B) = \underbrace{P(A \md B)}_{\tiny{\text{conditional probability}}}\cdot P(B) = P(B \md A)\cdot P(A)$$


* $\textit{Sum rule}: \text{disjunction of two events } A \text{ and } B$

$$P(A \cup B ) = P(A) + P(B) - P(A \cap B)$$


## $$\textbf{Marginalization (third rule)}$$
<hr>



* $\text{Discrete case:}$ 

$\text{     If events } B_1, \dots B_N \text{ are mutually exclusive (cannot occur at the same time) with } \sum_{i=1}^N P(B_i) = 1, \text{  then}$

$$P(A) = \sum_{i=1}^N P(A,B_i) = \sum_{i=1}^N
P(A|B_i)P(B_i)$$

* $\text{Continous case: } {\Large{\int}}_{\Omega} P(B)\> dB = 1$

$$P(A) = {{\int}}_{\Omega} P(A,B)\> dB = {{\int}}_{\Omega}
P(A|B)P(B)$$

## $$\textbf{Marginalization example}$$
<hr>


* $\text{Factory } X\text{’s bulbs work for over 5000 hours in
99% of cases. Event: $P(A|B_x)$ }$

* $\text{Factory } Y\text{’s bulbs work for over 5000 hours in 95% of cases. Event: $P(A|B_y)$ }$



$\text{It is known that factory} X \text{ supplies 60% of the total bulbs available (event $B_x$) and } Y \text{ supplies 40%  (event $B_y$).}$

<br>
$\textit{What is the chance that a purchased bulb will work for longer than 5000 hours (event A)?}$

* $\text{ Applying marginalization, we have: }$
<br>

$$P(A) = P(A \md B_x )P(B_x) + P(A \md B_y )P(B_y) = $$ 
<br>

$$ \frac{99}{100}\cdot \frac{6}{10} + \frac{95}{100}\cdot\frac{4}{10} =
\frac{594 + 380}{1000} =
\frac{974}{1000} = 97.4 \%$$


## $$\textbf{Bayes theorem}$$
<hr>

* $\text{Following the product rule  of probabilities:}$

$$ P(B \md A) \cdot P(A) = P(A \md B)\cdot P(B)\quad \Rightarrow $$

* $\text{Bayes theorem}$

$$\boxed{P(B \md A) = \frac{P(A \md B) \cdot P(B)}{P(A)}}$$



* $\text{Appying continous marginalization and product rule in the denominator:}$

$${P(B \md A) = \frac{P(A \md B)\cdot P(B)}{{\Large{\int}}_{\Omega} P(A,B)\> dB}} \Rightarrow$$

* $\text{Extended Bayes theorem}$

$$\boxed{P(B \md A) = \frac{P(A \md B) \cdot P(B)}{{\Large{\int}}_{\Omega} P(A \md B) \cdot P(B)\> dB}}$$


## $$\textbf{Bayes theorem}$$
<hr>

* $\text{Bayes’ Theorem is a simple mathematical formula},$ $\text{that describes how to update the probabilities of hypotheses when evidence is given.}$

* $\text{Given a hypothesis } H \text{ and evidence } E.$

$${\Large{\underbrace{P(H \md E)}_{\text{posterior}} = \frac{\overbrace{P(E \md H)}^{\text{likelihood}} \cdot \overbrace{P(H)}^{\text{prior}}}{\underbrace{P(E)}_{\text{evidence}}}}}$$

$\text{In other words}$:

$$\boxed{{\text{posterior} \propto \text{likelihood } \times \text{ prior}}}$$

## $$\textbf{Bayesian analysis}$$
<hr>

* $\text{It's about updating our beliefs in the light of new evidence:}$

    * $\t{starting with a prior belief (prior probability)}$
    
    
    * $\t{then incorporating new data (likelihood)}$ 
    
 
    * $\t{to form an updated belief (posterior probability)}$

## $$\textbf{Example task}$$
<hr>

* $\text{Given (observed) data } \mathcal{X} = \{ x_0, x_1, \dots, x_N\},\> \text{where } \mathcal{X} \in \mathbb{R}^{N}$

```python
import numpy as np
np.random.seed(42)

# ground truth prior distibutions
mu = np.random.normal(loc=0, scale=1)
sigma = 0.5 # Dirac distribution

print(f"mu: {mu:0.2f} std: {sigma}")

# likelihood ()dic
X = np.random.normal(loc=mu, scale=sigma, size=128) 
X[:10]
```

## $$\textbf{Example task cont'd}$$
<hr>


* $\text{Task is to find a distribution and its parameters,}$ $\text{which fits the observed data by appliying the Bayes theorem.}$

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
from seaborn import kdeplot

ax = plt.figure(figsize=(10, 6)).gca()

space = np.linspace(-1, 2, 1000)
ax.hist(X, bins=30, rwidth=0.9, alpha=0.5, density=True, label='Data')

ax.plot(space, norm.pdf(space, mu, sigma), linewidth=3, label='True distribution density')
kdeplot(X, ax=ax, label="Emprirical estimation", c='g', linestyle='--')

ax.legend(fontsize=16);
```

## $$\textbf{Bayesian modeling (example)}$$
<hr>


* $\textbf{Model distribution}  \text{ (likelihood) is assumed to be an univariate normal (Gaussian).}$

$$p(x_i \md \mu,\> \sigma) = \mathcal{N}(x_i \md \mu,\> \sigma )= \frac{1}{\sqrt{2\pi \sigma^2}}\exp\Big(-\frac{(x - \mu)^2}{2 \sigma^2}\Big)$$

* $\textbf{Parameters: } w = \{ \mu,\> \sigma \},\> \text{where } \mu \in \mathbb{R}, \sigma \in \mathbb{R}_{+}  $ 

* $\text{All data points assumed to be iid given the parameters.}$

$$p(\mathcal{X} \md w) = \prod_{i=1}^N p(x_i \md \mu,\> \sigma=1) = \prod_{i=1}^N \mathcal{N}(x_i \md \mu,\> \sigma=1)$$

* $\textbf{Prior distribution} \text{ (over parameters)} $

$$p(\mu) = \mathcal{N}(\mu \md \underbrace{m=0, s=1}_{\text{hyperparameters}}), \quad p(\sigma) = Dirac(\sigma \md \underbrace{m=0.5} _{\text{hyperparameter}})$$

## $$\textbf{Bayes posterior distribution}$$
<hr>
<br><br>
$${\Large{\underbrace{\overbrace{p(w \md \mX )}^{\color{red}{\text{posterior}}} \cdot \overbrace{p(\mX)}^{\color{magenta}{\text{marginal}}}}_{\color{red}{\text{must compute}}} = \underbrace{\overbrace{p(w,\mX)}^{\text{joint}} =  \overbrace{p(\mX \md w)}^{\color{blue}{\text{likelihood}}} \cdot \overbrace{p(w)}^{\color{blue}{\text{prior}}}}_{\color{green}{\text{given}}}}}$$

## $$\textbf{Marginal likelihood}$$
<hr>
<br>


* $\text{Normalization factor of a posterior distribution}$

$$p(\mX) = \int p(w,\, \mX )\, dw = \int p(\mX \md w )\cdot p(w) \, dw = $$

$$\prod_{i=1}^N \int p( x_i \md \mu, \, \sigma  ) \cdot p(\mu) \cdot p(\sigma)\, d\mu \, d\sigma = $$

$$ \prod_{i=1}^N \int \mathcal{N} ( x_i \md \mu, \, \sigma  ) \cdot \mathcal{N}(\mu | m=0, s=1) \cdot Dirac(\sigma \md m=1)\, d\mu \, d\sigma $$

## $$\textbf{Bayes estimator}$$
<hr>

* $\text{Expected value of the posterior distribution}$

$$\boxed{\Large\hat{w} = \mathbb{E}_{\small w \sim p(w \md \mX)}[w] = {\large{\int}} w \cdot p(w \md \mX) \, dw}$$

* $\text{Discrete example: Expected value of eyes throw for a perfect dice}$


$$\hat{x} = \mathbb{E}_{\small x \sim p(x \md \mX)}[x] = \sum_{n=1}^6 n \cdot \underbrace{p(n \md \mX)}_{\text{perfect dice}} = \frac{1}{6} \sum_{n=1}^N n = 3,5 $$

## $$\textbf{Predictive distribution}$$
<hr>



$${p(x_{new} \md \mX) = \int p(x_{new}, w \md \mX) \, dw} = {\underbrace{\int  p(x_{new}, \md w, \mX)}_{ x_{new}\, \indep \, \mX} \cdot \underbrace{p(w \md \mX)}_{\text{posterior}}\, dw} $$

$${p(x_{new} \md \mX) = \int \underbrace{p(x_{new} \md  w )}_{\text{likelihood}} \cdot \underbrace{p( w \md \mX)}_{\text{posterior}}\, dw} $$

```python
from scipy.stats import beta

a, b = .4, .3
samples = beta.rvs(a, b, size=100)
ax = plt.figure(figsize=(12, 6)).gca()
ax.hist(samples, bins=30, rwidth=.8, alpha=0.5, density=True, label='data')
kdeplot(samples, label='estimated density', ax=ax)
ax.legend(fontsize=16)
ax.grid()
```

## $$\textbf{Beta distribution}$$
<hr>

* $\text{Beta PDF:}\quad \alpha >0 \text{ and } \beta > 0 \text{ are shape parameters}$

$$p(x \md \alpha,\, \beta) = \frac{x^{\alpha-1}(1-x)^{\beta-1}}{\mathcal{B}(\alpha, \beta)}, \quad x \in (0,1)$$

$\text{where}$

$$\mathcal{B}(\alpha, \beta) = \frac{\Gamma(\alpha)\Gamma(\beta)}{\Gamma(\alpha + \beta)}$$

```python
from scipy.special import loggamma

def robust_beta_pdf(x, a, b):
    ln_B = loggamma(a) + loggamma(b) - loggamma(a + b)
    ln_ = (a - 1)*np.log(x) + (b - 1)*np.log(1 - x)
    return np.exp(ln_- ln_B)

np.allclose(robust_beta_pdf(rvs, a, b), beta.pdf(rvs, a, b))
```

## $$\textbf{Coin flipping posterior}$$
<hr>


$$\overbrace{p(\theta | h, \alpha, \beta)}^{\text{posterior}} =  \overbrace{\theta^{h} (1 - \theta)^{n-h}}^{\text{likelihood}} \overbrace{\theta^{\alpha-1}(1 - \theta)^{\beta-1}}^{\text{prior}} \cdot const
$$


* $\text{where}$ 

$$const = \frac{\binom{n}{h}}{B(\alpha,\beta)}$$

* $\text{after rearranging the term looks like a Beta distribution.}$

$$
\large{p(\theta \md h, \alpha, \beta) \propto \theta^{ h + \alpha -1 }(1 - \theta)^{n - h + \beta-1} }
$$

$$
\large{p(\theta \md h, \alpha, \beta) = \text{Beta}(\hat{\alpha},\, \hat{\beta}), \quad \text{where} \quad \hat{\alpha} = h+\alpha, \quad \hat{\beta} = n - h + \beta }
$$

## $$\textbf{Coin flipping}$$
<hr>

```python
beta_prior_params = [(1, 1), (0.1, 0.1), (10, 10)] # multiple a and b
```

```python
def posterior(h, n, a, b):
    a_hat = a + h
    b_hat = b + n - h
    return beta.pdf(space, a_hat, b_hat), a_hat, b_hat
```

```python
space = np.linspace(0, 1, 100)

fig = plt.figure(figsize=(20,20))
for idx, n in enumerate(ns):
    if idx == 0:
        plt.subplot(5, 3, 2)
    else:
        plt.subplot(5, 3, idx+3)
    h = heads[idx]
    for (a, b), c in zip(beta_prior_params, ('b', 'r', 'g')):
        post, a_hat, b_hat = posterior(h, n, a, b)
        plt.plot(space, post, c, label=f'a: {a} b: {b}')
        plt.grid(axis='x')

    plt.axvline(a_hat/(a_hat + b_hat), ymax=.8, c='r', linestyle='--')
    plt.axvline(theta_real, ymax=0.8, color='g', linestyle='--')
    plt.plot(0, 0, label="{:d} experiments\n{:d} heads".format(n,h), alpha=0)
    plt.xlim(0,1)
    plt.xlabel(r'$\theta$', fontsize=15)
    plt.legend(fontsize=14)
    plt.yticks([])
    plt.tight_layout()

E_value = a_hat/(a_hat + b_hat) # aka Bayes estimator 
print()
```

## $$\textbf{Conjugacy}$$
<hr>

* $\text{In the coin example the selection of the prior distribution}$
$\text{was not made arbitrary.}$



* $\text{In order to get a closed-form for your posterior you've to pick a distribution, which}$ $\text{depends on your model distribution.}$

* $\text{Such distribution is called conjugate.}$

* $\text{Then your posterior will have the same form as your conjugate prior.}$

$$\overbrace{Beta}^{\text{posterior}} = \overbrace{Bernoulli}^{\text{model (likelihood)}} \cdot \overbrace{Beta}^{\text{conj. prior}}$$
